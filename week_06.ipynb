## nltk corpus`

```python
import nltk
nltk.download('brown')
```

```python
brown=nltk.corpus.brown
```

```python
#복습
cfd1=nltk.ConditionalFreqDist(
    (genre, word)
    for genre in brown.categories()
    for word in brown.words(categories=genre)
)
genres = brown.categories()[:5]
modals = ['can', 'may', 'could', 'would', 'will', 'might']
cfd1.tabulate(conditions=brown.categories(), samples=modals)
```

```python
cfd2=nltk.ConditionalFreqDist(
    (genre, modals)
        for genre in brown.categories()
        for modals in ['can','could','may', 'would', 'will', 'might']
        for word in brown.words(categories=genre)
            if word.lower() == modals
)
cfd2.tabulate()
```

### 순회하기
- zip() : 여러개 리스트 동시 순회
- enumerate() : 인덱스도 함께 들고자 할때(list, range)
- range()

```python
a = [1,2,3]
b= ['mon', 'tue','wed','thu']
for itemA, itemB in zip(a, b): #아이템 수가 적은 시퀀스에서 완료
    print((itemA, itemB))
```

```python
list(zip(a,b))
```

```python
dict(zip(a,b))
```

```python
for idx, itemA in enumerate(b):
    print((idx, itemA))
```
